```python
# Importing all the required library for the project.
import pandas as pd
import re
import selenium 
from selenium import webdriver
from time import sleep
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys

import warnings
warnings.filterwarnings('ignore')
```

```python
driver= webdriver.Chrome("chromedriver.exe")
# Get the webpage
url="https://www.amazon.in/" 
driver.get(url)
#driver.maximize_window()
```

```python
search_product=driver.find_element_by_xpath('//input[@type="text"]')
search_product.send_keys("Sarees")
```

```python
search_btn=driver.find_element_by_xpath('(//input[@type="submit"])[1]')
search_btn.click()
```

```python
Sarees=[]
elements = driver.find_elements_by_xpath("//img[@class='s-image']")
for i in range(5):
    print('Scraping Page',i+1)
    for i in elements:
        image=i.get_attribute('src')
        Sarees.append(image)
```
